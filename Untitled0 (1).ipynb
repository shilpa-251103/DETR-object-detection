```python
%%writefile /content/drive/MyDrive/Object-Detection-Using-DETR-CustomDataset/datasets/face_to_coco.py
import re
import os
import cv2
import json
import itertools
import numpy as np
from glob import glob
import scipy.io as sio
from pycocotools import mask as cocomask
from PIL import Image

MAX_N = 10

categories = [
    {
        "supercategory": "none",
        "name": "face",
        "id": 0
    }
]
phases = ["train", "val"]

for phase in phases:
    root_path = "/content/drive/MyDrive/Object-Detection-Using-DETR-CustomDataset/datasets/WIDER_{}/images/".format(phase)
    gt_path = os.path.join("/content/drive/MyDrive/Object-Detection-Using-DETR-CustomDataset/datasets/wider_face_split/wider_face_{}.mat".format(phase))
    json_file = "/content/drive/MyDrive/Object-Detection-Using-DETR-CustomDataset/datasets/WIDER_{}/{}.json".format(phase, phase)

    print(f"Processing phase: {phase}")
    print(f"Ground truth path: {gt_path}")
    print(f"Output JSON file: {json_file}")

    gt = sio.loadmat(gt_path)
    event_list = gt.get("event_list")
    file_list = gt.get("file_list")
    face_bbox_list = gt.get("face_bbx_list")

    res_file = {
        "categories": categories,
        "images": [],
        "annotations": []
    }
    annot_count = 0
    image_id = 0
    processed = 0
    for event_idx, path in enumerate(event_list):
        base_path = path[0][0]
        for file_idx, img_name in enumerate(file_list[event_idx][0]):
            file_path = img_name[0][0]
            face_bbox = face_bbox_list[event_idx][0][file_idx][0]
            num_boxes = face_bbox.shape[0]

            if num_boxes > MAX_N:
                continue

            img_path = os.path.join(root_path, base_path, file_path + ".jpg")
            filename = os.path.join(base_path, file_path + ".jpg")
            if not os.path.exists(img_path):
                print(f"Image path does not exist: {img_path}")
                continue

            try:
                img = Image.open(img_path)
            except Exception as e:
                print(f"Error opening image {img_path}: {e}")
                continue

            img_w, img_h = img.size
            img_elem = {"file_name": filename,
                        "height": img_h,
                        "width": img_w,
                        "id": image_id}

            res_file["images"].append(img_elem)

            for i in range(num_boxes):
                xmin = int(face_bbox[i][0])
                ymin = int(face_bbox[i][1])
                xmax = int(face_bbox[i][2]) + xmin
                ymax = int(face_bbox[i][3]) + ymin
                w = xmax - xmin
                h = ymax - ymin
                area = w * h
                poly = [[xmin, ymin],
                        [xmax, ymin],
                        [xmax, ymax],
                        [xmin, ymax]]
                annot_elem = {
                    "id": annot_count,
                    "bbox": [
                        float(xmin),
                        float(ymin),
                        float(w),
                        float(h)
                    ],
                    "segmentation": list([poly]),
                    "image_id": image_id,
                    "ignore": 0,
                    "category_id": 0,
                    "iscrowd": 0,
                    "area": float(area)
                }

                res_file["annotations"].append(annot_elem)
                annot_count += 1

            image_id += 1

            processed += 1

    with open(json_file, "w") as f:
        json_str = json.dumps(res_file)
        f.write(json_str)

    print(f"Processed {processed} {phase} images...")
print("Done.")
```

```python
import torch
import torchvision.transforms as T
import matplotlib.pyplot as plt
from PIL import Image
import time
import math
import os
```

```python
CLASSES = ['face']

# colors for visualization
COLORS = [[0.000, 0.447, 0.741]]
```

```python
# Transformation
transform = T.Compose([
    T.Resize(800),
    T.ToTensor(),
    T.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])
```

```python
# Box conversion functions
def box_cxcywh_to_xyxy(x):
    x_c, y_c, w, h = x.unbind(1)
    b = [(x_c - 0.5 * w), (y_c - 0.5 * h),
         (x_c + 0.5 * w), (y_c + 0.5 * h)]
    return torch.stack(b, dim=1)

def rescale_bboxes(out_bbox, size):
    img_w, img_h = size
    b = box_cxcywh_to_xyxy(out_bbox)
    b = b * torch.tensor([img_w, img_h, img_w, img_h], dtype=torch.float32)
    return b
```

```python
# Function to plot results and save image
def plot_results_and_save(pil_img, prob, boxes, save_path):
    plt.figure(figsize=(16, 10))
    plt.imshow(pil_img)
    ax = plt.gca()
    colors = COLORS * 100
    for p, (xmin, ymin, xmax, ymax), c in zip(prob, boxes.tolist(), colors):
        ax.add_patch(plt.Rectangle((xmin, ymin), xmax - xmin, ymax - ymin,
                                   fill=False, color=c, linewidth=3))
        cl = p.argmax()
        text = f'{CLASSES[cl]}: {p[cl]:0.2f}'
        ax.text(xmin, ymin, text, fontsize=15,
                bbox=dict(facecolor='yellow', alpha=0.5))
    plt.axis('off')
    plt.savefig(save_path, bbox_inches='tight')
    plt.close()
```

```python
# Post-process image function
def postprocess_img(img_path, save_path):
    im = Image.open(img_path)
    img = transform(im).unsqueeze(0)
    start = time.time()
    outputs = model(img)
    end = time.time()
    print(f'Prediction time per image: {math.ceil(end - start)}s')

    probas = outputs['pred_logits'].softmax(-1)[0, :, :-1]
    keep = probas.max(-1).values > 0.9
    bboxes_scaled = rescale_bboxes(outputs['pred_boxes'][0, keep], im.size)

    plot_results_and_save(im, probas[keep], bboxes_scaled, save_path)
```

```python
# Main loop to process images
def postprocess_and_save(image_path, output_dir):
    image_name = os.path.basename(image_path)
    save_path = os.path.join(output_dir, image_name)
    postprocess_img(image_path, save_path)
    print(f"Image saved to {save_path}")
```

```python
import re
import pandas as pd
import matplotlib.pyplot as plt

# Path to the log file
log_file_path = '/content/drive/MyDrive/Object-Detection-Using-DETR-CustomDataset/detr/outputs/log.txt'

# Initialize lists to store data
epochs = []
train_losses = []
val_losses = []

# Regular expressions to match the data
train_loss_pattern = re.compile(r'"train_loss": ([0-9.]+)')
val_loss_pattern = re.compile(r'"test_loss": ([0-9.]+)')
epoch_pattern = re.compile(r'"epoch": (\d+)')

# Variables to hold current values
current_epoch = None
train_loss = None
val_loss = None

# Read and parse the log file
with open(log_file_path, 'r') as file:
    for line in file:
        # Match epoch
        epoch_match = epoch_pattern.search(line)
        if epoch_match:
            current_epoch = int(epoch_match.group(1))
            epochs.append(current_epoch)

        # Match training loss
        train_loss_match = train_loss_pattern.search(line)
        if train_loss_match:
            train_loss = float(train_loss_match.group(1))
            train_losses.append(train_loss)

        # Match validation loss
        val_loss_match = val_loss_pattern.search(line)
        if val_loss_match:
            val_loss = float(val_loss_match.group(1))
            val_losses.append(val_loss)

# Check if data was successfully extracted
if not epochs or not train_losses or not val_losses:
    print("No data found. Please check the log file and patterns.")
else:
    # Create a DataFrame for plotting
    df = pd.DataFrame({
        'Epoch': epochs,
        'Train Loss': train_losses,
        'Validation Loss': val_losses
    })

    # Save DataFrame to CSV for debugging (optional)
    df.to_csv('/content/drive/MyDrive/Object-Detection-Using-DETR-CustomDataset/detr/outputs/losses.csv', index=False)

    # Plotting
    plt.figure(figsize=(10, 5))
    plt.plot(df['Epoch'], df['Train Loss'], label='Training Loss', marker='o', linestyle='-')
    plt.plot(df['Epoch'], df['Validation Loss'], label='Validation Loss', marker='o', linestyle='-')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Training and Validation Loss Curves')
    plt.legend()
    plt.grid(True)
    plt.savefig('/content/drive/MyDrive/Object-Detection-Using-DETR-CustomDataset/detr/outputs/loss_curves.png')  # Save the plot
    plt.show()
```
